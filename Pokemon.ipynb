```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql
```

```python
#Extract CSVs into DataFrames
```

```python
pokemon_dataset_A =  "../ETL/Resources/pokemon.csv"
pokemon_dataset_A_df = pd.read_csv(pokemon_dataset_A)
pokemon_dataset_A_df.head()
```

```python
pokemon_dataset_B =  "../ETL/Resources/pokemon_alopez247.csv"
pokemon_dataset_B_df = pd.read_csv(pokemon_dataset_B)
pokemon_dataset_B_df.head()
```

```python
#Transform Pokemon Dataset A DataFrame
```

```python
#extract the columns that dataset B not have
new_pokemon_dataset_A_df = pokemon_dataset_A_df[['id', 'ndex', 'species','forme', 'abilityH','dex1', 'dex2',
                                                 'class', 'percent-female', 'pre-evolution']].copy()
new_pokemon_dataset_A_df.head()
```

```python
# Rename Poke new_pokemon_dataset_A columns
pokemon_A = new_pokemon_dataset_A_df.rename(index=str, columns={'ndex':'pokedex_no', 'species':'P_name','abilityH':'hidden_ability',
                                                                'dex1':'description1', 'dex2':'description2',
                                                                'percent-female':'percent_female'})

pokemon_A = pokemon_A.drop_duplicates("id")
pokemon_A = pokemon_A.set_index("id")
pokemon_A.head()
```

```python
#save Pokemon Dataset A as a CSV file
pokemon_A.to_csv('Pokemon_A.csv')
```

```python
# Rename pokemon_dataset_B_df columns
pokemon_B = pokemon_dataset_B_df.rename(index=str, columns={'Number':'id', 'Name':'P_name'})
pokemon_B = pokemon_B.drop_duplicates("id")
pokemon_B = pokemon_B.set_index("id")
pokemon_B.head()
```

```python
#save New Pokemon Dataset B as a CSV file
pokemon_B.to_csv('Pokemon_B.csv')
```
